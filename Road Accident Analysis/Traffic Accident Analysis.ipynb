```python
df['hour'] = df['hrmn'] // 100   # divided hrmn col by 1000 to get hours
df.drop(columns=['hrmn'], inplace=True)
```

```python
#feature engineering
df['age_gap'] = df['driver_age'] - df['car_age']# Then check correlation of 'age_gap' with your target
# Driver age group
df['driver_age_group'] = pd.cut(
    df['driver_age'],
    bins=[18, 25, 60, 71],  # max 70 tak, isliye last bin 71
    labels=['Young', 'Middle', 'Senior'],
    right=False
)
# Casualty age group
df['casualty_age_group'] = pd.cut(
    df['casualty_age'],
    bins=[1, 18, 60, 81],  # max 80 tak, isliye last bin 81
    labels=['Minor', 'Adult', 'Senior'],
    right=False
)
# Relative age difference
df['relative_age'] = df['casualty_age'] - df['driver_age']
```

```python
# Accident-specific time-of-day buckets
df['time_of_day'] = pd.cut(
    df['hour'],
    bins=[0, 5, 8, 12, 17, 21, 24],  # 24-hour format boundaries
    labels=[
        'Late Night',       # 0 - 5 : very less traffic
        'Morning Rush',     # 5 - 8 : office/school rush
        'Midday',           # 8 - 12 : stable traffic
        'Afternoon',        # 12 - 17 : mixed traffic
        'Evening Rush',     # 17 - 21 : peak rush hour
        'Night'             # 21 - 24 : traffic drop 
    ],
    right=False,
    ordered=False
)
df['engine_size_cat'] = pd.cut(
    df['engine_size'], 
    [0, 1000, 2000, np.inf], 
    labels=['Small', 'Medium', 'Large'], 
    right=False
)

bad_weather_terms = ['rain', 'snow', 'fog', 'storm'] #gropued rain,snow,fog and storm as is_bad weather
df['is_bad_weather'] = df['weather'].str.lower().apply(lambda x: int(any(t in x for t in bad_weather_terms)))
df['is_dark'] = df['lum'].str.lower().apply(lambda x: int('night' in x or 'dark' in x)) #if 'night' or 'dark' is present it considers under is_dark if it is dark it is 1 else 0
```

```python
#drop columns driver_age,car_age,casualty_age,weather and lum
df.drop(columns=['driver_age'],inplace=True)
df.drop(columns=['car_age'],inplace=True)
df.drop(columns=['casualty_age'],inplace=True)
df.drop(columns=['weather','lum'],inplace=True)
df.drop(columns=['engine_size'],inplace=True)
```
